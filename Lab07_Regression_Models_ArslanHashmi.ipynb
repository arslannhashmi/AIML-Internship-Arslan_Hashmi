## Dataset Overview: California Housing Dataset

In this lab, we will use the **California Housing Dataset** (available directly in `scikit-learn`).

- **Target variable (`y`)**: `MedHouseVal` — Median house value for California districts (expressed in hundreds of thousands of dollars, e.g., 2.0 = $200,000).
- **Features (`X`)**:
  1. `MedInc`: Median income in block group
  2. `HouseAge`: Median house age in block group
  3. `AveRooms`: Average number of rooms per household
  4. `AveBedrms`: Average number of bedrooms per household
  5. `Population`: Block group population
  6. `AveOccup`: Average number of household members
  7. `Latitude`: Block group latitude
  8. `Longitude`: Block group longitude

## Example 1: Loading & Exploring the Dataset

First, let's load the dataset, create a Pandas DataFrame, and inspect the target variable and features.

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np

# Load California Housing dataset
housing = fetch_california_housing(as_frame=True)
df = housing.frame

print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nFeature names:", housing.feature_names)
print("\nTarget name:", housing.target_names)
```

## Example 2: Simple Linear Regression & Model Evaluation

Let's predict house prices (`MedHouseVal`) using just one feature: **Median Income (`MedInc`)**. 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Select single feature and target
X_simple = df[['MedInc']]
y = df['MedHouseVal']

# Train-test split (80-20)
X_train, X_test, y_train, y_test = train_test_split(X_simple, y, test_size=0.2, random_state=42)

# Train model
model_simple = LinearRegression()
model_simple.fit(X_train, y_train)

# Predictions
y_pred_simple = model_simple.predict(X_test)

# Evaluation
mse = mean_squared_error(y_test, y_pred_simple)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred_simple)

print(f"Slope (Coefficient): {model_simple.coef_[0]:.4f}")
print(f"Intercept: {model_simple.intercept_:.4f}")
print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R2 Score: {r2:.4f}")
```

## Example 3: Multiple Linear Regression

Now, let's use multiple features (`MedInc`, `HouseAge`, `AveRooms`) to improve our prediction model.

```python
# Select multiple features
features = ['MedInc', 'HouseAge', 'AveRooms']
X_multi = df[features]

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(X_multi, y, test_size=0.2, random_state=42)

# Train multiple linear regression
model_multi = LinearRegression()
model_multi.fit(X_train_m, y_train_m)

y_pred_multi = model_multi.predict(X_test_m)

# Evaluate
rmse_multi = np.sqrt(mean_squared_error(y_test_m, y_pred_multi))
r2_multi = r2_score(y_test_m, y_pred_multi)

print(f"Multiple Regression RMSE: {rmse_multi:.4f}")
print(f"Multiple Regression R2 Score: {r2_multi:.4f}")
print("\nFeature Coefficients:")
for feature, coef in zip(features, model_multi.coef_):
    print(f"{feature}: {coef:.4f}")
```

### Task 1 — Full Feature Regression

```python
# Task 1: Full Feature Regression

# Use all 8 features
X_full = df.drop(columns=['MedHouseVal'])
y = df['MedHouseVal']

# 80-20 train-test split
X_train_full, X_test_full, y_train_full, y_test_full = train_test_split(
    X_full, y, test_size=0.2, random_state=42
)

# Train Linear Regression on all features
model_full = LinearRegression()
model_full.fit(X_train_full, y_train_full)

# Predictions
y_pred_full = model_full.predict(X_test_full)

# Evaluation
rmse_full = np.sqrt(mean_squared_error(y_test_full, y_pred_full))
r2_full = r2_score(y_test_full, y_pred_full)

print("=== Full Feature Linear Regression ===")
print(f"Number of features: {X_full.shape[1]}")
print(f"RMSE: {rmse_full:.4f}")
print(f"R² Score: {r2_full:.4f}")
```

### Task 2 — Coefficient Analysis

```python
# Task 2: Coefficient Analysis

print("=== Feature Coefficients ===")
coef_df = pd.DataFrame({
    'Feature': X_full.columns,
    'Coefficient': model_full.coef_
}).sort_values(by='Coefficient', ascending=False)

print(coef_df.to_string(index=False))

# Identify strongest positive impact
strongest_feature = coef_df.iloc[0]['Feature']
strongest_coef = coef_df.iloc[0]['Coefficient']

print(f"\nStrongest positive impact feature: {strongest_feature}")
print(f"Coefficient value: {strongest_coef:.4f}")
print("→ Higher median income is the strongest driver of higher house values.")
```

### Task 3 — Comparison

The simple linear regression model (using only `MedInc`) achieved an R² score of approximately **0.46**, while the full-feature model (using all 8 features) improved to around **0.58**. Including additional features such as house age, average rooms, location (latitude/longitude), and population provides the model with richer information about the factors that influence house prices. This multi-dimensional view allows the model to capture more of the variance in the target variable, resulting in better predictive performance and a higher R² score.

### Task 4 — Reflection

In this lab I learned how to build and evaluate regression models using scikit-learn, starting from a simple one-feature model and progressing to multiple linear and polynomial regression. The key insight was that evaluation metrics serve different purposes: **RMSE** tells us the typical size of prediction errors in the original units (easier to interpret in dollars), while **R²** tells us the proportion of variance explained by the model (useful for comparing models).
